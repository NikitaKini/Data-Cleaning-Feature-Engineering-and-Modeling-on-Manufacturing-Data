```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle

x = pd.read_csv(r'C:\Users\Rexroth\Nikita\Final_code_with_GUI\Automation_curr\Feature_without figures_only value 2.csv', encoding = 'ISO-8859-1', sep=';')
y = pd.read_csv(r'C:\Users\Rexroth\Nikita\Final_code_with_GUI\Automation_curr\Feature_without figures_only value 2.csv', encoding = 'ISO-8859-1', sep=';')

x = x.T[1:9].T #we only want to consider specific / selected features 
y = y.T[10:11].T #we only want to consider the diameter  
# X = x

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state=8)

from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=100, max_features='sqrt', criterion='mae', min_samples_split=4)

#Training the model on basis of the X_train data 
rfr.fit(X_train, y_train)
predictions = rfr.predict(X_test)
R_quadrat = rfr.score(X_test, y_test)

#Depicting the performance of the model on basis of the X_test data
print(predictions)
print(R_quadrat)

# #Saving the trained model
filename = '17010847_finalized_trained_RFR_model_for_diameter_prediction.sav'
# print('valve',filename.split('_')[0])
pickle.dump(rfr, open(filename, 'wb'))

#https://machinelearningmastery.com/save-load-machine-learning-models-python-scikit-learn/
```
